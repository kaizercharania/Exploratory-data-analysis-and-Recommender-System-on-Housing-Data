```python
import requests
import pandas
import re
from bs4 import BeautifulSoup
import lxml
import time
myfile = open("/Users/kaizer/Desktop/cities.txt",encoding='utf-8')
data1 = myfile.read()
myfile.close()
data1 = list(data1.splitlines())
search_list = []
i=0
s = 1
data = []
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.90 Safari/537.36'}
for i in range(len(data1)):
    string_name = data1[i]+"-ca/" 
    string_name = string_name.replace(" ","-")
    search_list.append(string_name)
for i in range(len(search_list)):
    time.sleep(10)
    url = "https://www.zillow.com/" + str(search_list[i])
    page_requests =  requests.get(url,headers=headers)
    page_content = page_requests.content
    page_soup = BeautifulSoup(page_content,"html5lib")
    #print(page_soup)
    for li in page_soup.find_all('div', {'class': 'zsg-photo-card-caption'}):
        #print(li)
        i +=1
        print(i)
        try:
            if li.find('span', {'class': 'zsg-photo-card-price'}) !=None:
                price = (li.find('span', {'class': 'zsg-photo-card-price'}).text)
            if li.find('span', {'class': 'zsg-photo-card-info'}) != None:
                info = (li.find('span', {'class': 'zsg-photo-card-info'}).text)
            if li.find('span', {'class': 'zsg-photo-card-address'}) !=None:
                address = (li.find('span', {'class': 'zsg-photo-card-address'}).text)
            #print(li.find('span', {'class': 'zsg-photo-card-broker-name'}).text)
            d = {"an0": s ,"price":price,"info":info,"address":address}
            s +=1
            data.append(d)
        except :
            print('An error occured')
    
    while page_soup.find('link',{"rel":"next"}) != None:
        time.sleep(10)
        try:
            page_requests =  requests.get(url,headers=headers)
            page_content = page_requests.content
            page_soup = BeautifulSoup(page_content,"html5lib")
            url = page_soup.find('link',{"rel":"next"}, href=True)['href']
            print(url)
            for li in page_soup.find_all('div', {'class': 'zsg-photo-card-caption'}):
                #print(li)
                i +=1
                print(i)
                try:
                    if li.find('span', {'class': 'zsg-photo-card-price'}) !=None:
                        price = (li.find('span', {'class': 'zsg-photo-card-price'}).text)
                    if li.find('span', {'class': 'zsg-photo-card-info'}) != None:
                        info = (li.find('span', {'class': 'zsg-photo-card-info'}).text)
                    if li.find('span', {'class': 'zsg-photo-card-address'}) !=None:
                        address = (li.find('span', {'class': 'zsg-photo-card-address'}).text)
                    #print(li.find('span', {'class': 'zsg-photo-card-broker-name'}).text
                    d = {"an0": s ,"price":price,"info":info,"address":address}
                    s +=1
                    data.append(d)
                except :
                    print('An error occured')
        except:
            pass
        df = pandas.DataFrame(data)
        df.to_csv("data_zillow_dot_com.csv")
```
